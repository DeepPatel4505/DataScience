```python
import pandas as pd
import numpy as np
```

```python
path = "/sample_data/california_housing"
df = pd.read_csv(path)
df.head(10)
```

```python
X = df["X"]
Y = df["Y"]
theta0 = 0.4
theta1 = 0.4
alpha = 0.01
m = len(df)
```

```python
y_cap = theta0 + theta1 * X
# y_cap
(y_cap - Y)
```

```python
J = ((y_cap - Y) ** 2).sum() / (2 * m)
J
```

```python
deltaJ0 = (y_cap - Y).sum() / m
deltaJ1 = ((y_cap - Y) * X).sum() / m
deltaJ0 , deltaJ1
```

```python
t0 = theta0 - alpha * deltaJ0
t1 = theta1 - alpha * deltaJ1
t0 , t1
```

```python
data = pd.read_csv("Housing.csv")
data.sample(10)
```

```python
X = data.drop(['price'],axis=1)
X = pd.DataFrame(X)

Y = data["price"]
X.sample(10)
#  airconditioning , basement , prefarea
X["furnishingstatus"].value_counts()
Y.sample(10)
```

```python

from sklearn.preprocessing import LabelEncoder,StandardScaler

le = LabelEncoder()
st = StandardScaler()
cols = ["furnishingstatus" , "prefarea" , "basement" , "airconditioning"]

for col in cols:
  X[col] = le.fit_transform(X[col])

X = X.drop(["mainroad","guestroom","hotwaterheating"],axis=1)

st.fit(X)
X_scaled= st.fit_transform(X)
# Y_scaled = st.tranform(Y)

X = pd.DataFrame(X_scaled,columns=X.columns)
X
```

```python
X.describe()
```

```python
X["temp"] = 1
X.sample(10)
```

```python
theta = [x*0.1 for x in range(len(X.columns))]
theta
```

```python
from ctypes import Array
J_prev = float('inf')
tolerance = 1e-20
itr = 0
# print(theta0,theta1)


while True :
    itr+=1
    y_cap = theta0 + theta1 * X


    J = ((y_cap - Y) ** 2).sum() / (2 * m)

    if abs(J_prev - J) < tolerance or itr >= 1000:
        break

    deltaJ0 = (y_cap - Y).sum() / m
    deltaJ1 = ((y_cap - Y) * X).sum() / m

    t0 = theta0 - alpha * deltaJ0
    t1 = theta1 - alpha * deltaJ1

    theta0 = t0
    theta1 = t1

    J_prev = J

theta0, theta1, itr
```
