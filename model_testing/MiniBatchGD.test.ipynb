```python
from sklearn.datasets import make_regression
from matplotlib import pyplot as plt
```

```python
x,y = make_regression(n_samples=100,n_features=10,n_informative=1,n_targets=1,noise=20,random_state=43)
x.shape,y.shape,x[0].shape
```

```python
plt.scatter(x.T[0],y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
y_train[0]
```

```python
from models import MiniBatchGD
```

```python
model = MiniBatchGD(batch_size=5)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
model.coeff_,model.intercept_
```

```python
from sklearn.linear_model import LinearRegression
```

```python
mdl = LinearRegression()
mdl.fit(x_train,y_train)
```

```python
ypred = mdl.predict(x_test)
r2_score(y_test,ypred)
```
