```python
from sklearn.datasets import make_regression
from matplotlib import pyplot as plt
```

```python
x,y = make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20,random_state=43)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
from models import SimpleLR
```

```python
lr = SimpleLR()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
mse, r2
```

```python
from sklearn.linear_model import LinearRegression
mdl = LinearRegression()
mdl.fit(x_train,y_train)
y_pred_skl = mdl.predict(x_test)
mse_skl = mean_squared_error(y_test,y_pred_skl)
r2_skl = r2_score(y_test,y_pred_skl)
mse_skl, r2_skl
```
